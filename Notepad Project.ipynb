```python
import tkinter 
import os   
import time
from tkinter import *
from tkinter.messagebox import *
from tkinter import filedialog 
from tkinter import font
from tkinter.ttk import *
```

```python
def findfunc(txt):
    global textarea
    textarea.tag_remove("match","1.0",tkinter.END)
    matches=0
    if txt:
        start_pos="1.0"
        
        while True:
            start_pos= textarea.search(txt,start_pos,stopindex=tkinter.END)
            if not start_pos:
                break;
            end_pos=f"{start_pos}+{len(txt)}c"
            textarea.tag_add("match",start_pos,end_pos)
            matches+=1
            start_pos=end_pos
            textarea.tag_config('match',foreground="white",background="blue")    
            
```

```python
def replace(word,replace_text):
    global textarea
    content=textarea.get(1.0,tkinter.END)
    new_content=content.replace(word,replace_text)
    textarea.delete(1.0,tkinter.END)
    textarea.insert(1.0,new_content)
```

```python
def findwindow():
    def func1():
        txt=t.get()
        findfunc(txt)
        
    def cancel():
        txt=t.get()
        replace(txt,txt)
        findwin.destroy()
        
    findwin=tkinter.Tk()
    findwin.title("Find")
    findwin.geometry("430x90")
    tkinter.Label(findwin,text="Find what:").place(x = 2,y = 10)
    t=tkinter.Entry(findwin,width=35,bd=1)
    t.place(x=80,y=12)
    t.bind('<Return>', func1)
    t.bind('<Return>', cancel)
    
    
    tkinter.Button(findwin,width=12, height=1, text ="Find Next", command=func1).place(x=320,y=9)
    tkinter.Button(findwin,width=12, height=1, text ="Cancel", command=cancel).place(x=320,y=45)
    findwin.mainloop()
```

```python
def find_and_replacewindow():
    def findreplace():
        word=f.get()
        replace_text=r.get()
        replace(word,replace_text)
    def func2():
        txt=f.get()
        findfunc(txt)
    findwin=tkinter.Tk()
    findwin.title("Find and Replace")
    findwin.geometry("430x160")
    tkinter.Label(findwin,text="Find what:").place(x = 2,y = 10)
    tkinter.Label(findwin,text="Replace with:").place(x = 2,y = 45)
    f=tkinter.Entry(findwin,width=35,bd=1)
    f.place(x=80,y=12)
    f.bind('<Return>',findreplace)
    r=tkinter.Entry(findwin,width=35,bd=1)
    r.place(x=81,y=47)
    r.bind('<Return>',findreplace)
    tkinter.Button(findwin,width=12, height=1, text ="Find Next",command=func2).place(x=320,y=9)
    tkinter.Button(findwin,width=12, height=1, text ="Replace All", command=findreplace).place(x=320,y=44)
    tkinter.Button(findwin,width=12, height=1, text ="Cancel", command=findwin.destroy).place(x=320,y=79)
    findwin.mainloop()

```

```python

def Algerian():
    global textarea
    textarea.config(font="Algerian")  
def Arial():
    global textarea
    textarea.config(font="Arial")  
def Courier():
    global textarea
    textarea.config(font="Courier")  
def Cambria():
    global textarea
    textarea.config(font="Cambria")

def make_bold():
    global textarea
    current_tags = textarea.tag_names("sel.first")
    if "bt" in current_tags:
        textarea.tag_remove("bt", "sel.first", "sel.last")
    else:
        textarea.tag_add("bt", "sel.first", "sel.last")
        
def make_italic():
    global textarea
    current_tags = textarea.tag_names("sel.first")
    if "it" in current_tags:
        textarea.tag_remove("it", "sel.first", "sel.last")
    else:
        textarea.tag_add("it", "sel.first", "sel.last")
        
def make_underline():
    global textarea
    current_tags = textarea.tag_names("sel.first")
    if "ut" in current_tags:
        textarea.tag_remove("ut", "sel.first", "sel.last")
    else:
        textarea.tag_add("ut", "sel.first", "sel.last")
```

```python
# File Menu Functions
def newDoc(): 
    global textarea,file
    root.title("Untitled - Notepad") 
    file = None
    textarea.delete(1.0,END)
def openDoc(): 
    global textarea,file     
    file = filedialog.askopenfilename(defaultextension=".txt",filetypes=[("All Files","*.*"), ("Text Documents","*.txt")]) 
    if file == "": 
        return
    else: 
        root.title(os.path.basename(file) + " - Notepad") 
        textarea.delete(1.0,END) 
  
        location = open(file,"r") 
  
        textarea.insert(1.0,location.read()) 
  
        location.close()

def saveDoc():
    global textarea,file
    text=textarea.get("1.0","end-1c")
    if file==None:
        file=filedialog.asksaveasfilename(initialfile='Untitled.txt', defaultextension=".txt", filetypes=[("All Files","*.*"), ("Text Documents","*.txt")])
        
        
        if file=="":
            file=None
        else:
            location=open(file,"w+")
            location.write(text)
            location.close()
            root.title(os.path.basename(file) + " - Notepad")
    else:
        location=open(file,"w")
        location.write(text)
        location.close()
    
def Exit():
    root.destroy()
```

```python
def cut(): 
    global textarea
    textarea.event_generate("<<Cut>>") 
    
def copy(): 
    global textarea
    textarea.event_generate("<<Copy>>") 
    
def paste(): 
    global textarea
    textarea.event_generate("<<Paste>>")
```

```python
root=tkinter.Tk()
root.title("Notepad")

file=None

textarea=tkinter.Text(root,undo=True,font=("Georgia", "12"))
textarea.grid(sticky=N+S+E+W)
menubar=Menu(root)
root.config(menu=menubar)
fileMenu=Menu(menubar,tearoff=0)
fileMenu.add_command(label="New",command=newDoc)
fileMenu.add_command(label="Open",command=openDoc)
fileMenu.add_command(label="Save",command=saveDoc)
fileMenu.add_separator()
fileMenu.add_command(label="Exit",command=Exit)
menubar.add_cascade(label="File",menu=fileMenu)

editMenu=Menu(menubar,tearoff=0)
editMenu.add_command(label="Undo",command=textarea.edit_undo)
editMenu.add_command(label="Redo",command=textarea.edit_redo)
editMenu.add_separator()
editMenu.add_command(label="Cut",command=cut)
editMenu.add_command(label="Copy",command=copy)
editMenu.add_command(label="Paste",command=paste)
editMenu.add_separator()
editMenu.add_command(label="Find",command=findwindow)
editMenu.add_command(label="Find Next",command=findwindow)
editMenu.add_command(label="Fint and Replace",command=find_and_replacewindow)
menubar.add_cascade(label="Edit",menu=editMenu)

fontMenu=Menu(menubar,tearoff=0)

fontstyleMenu=Menu(fontMenu,tearoff=0)
fontstyleMenu.add_command(label="Bold",command=make_bold)
bold_font = font.Font(textarea, textarea.cget("font"))
bold_font.configure(weight="bold")
textarea.tag_configure("bt", font=bold_font)

fontstyleMenu.add_command(label="Italic",command=make_italic)
ita_font = font.Font(textarea, textarea.cget("font"))
ita_font.configure(slant="italic")
textarea.tag_configure("it", font=ita_font)

fontstyleMenu.add_command(label="Underline",command=make_underline)
under_font = font.Font(textarea, textarea.cget("font"))
under_font.configure(underline=True)
textarea.tag_configure("ut", font=under_font)

fontTypeMenu=Menu(fontMenu,tearoff=0)
fontTypeMenu.add_command(label="Arial",command=Arial)
fontTypeMenu.add_command(label="Algerian",command=Algerian)
fontTypeMenu.add_command(label="Cambria",command=Cambria)
fontTypeMenu.add_command(label="Courier",command=Courier)

fontMenu.add_cascade(label="Font Type",menu=fontTypeMenu)
fontMenu.add_cascade(label="Font Style",menu=fontstyleMenu)

menubar.add_cascade(label="Font",menu=fontMenu)

helpMenu=Menu(menubar,tearoff=0)
helpMenu.add_command(label="About Notepad",command="")
menubar.add_cascade(label="Help",menu=helpMenu)

root.mainloop()
```
